```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# CSV 파일 불러오기
df = pd.read_csv("./data/train/exp_1_DAY2/DAY2_exp1.csv")

# 샘플링 주파수 설정
fs = 200
n = len(df)

# 각 축별로 FFT 및 시각화
for axis in ['gyro_x', 'gyro_y', 'gyro_z']:
    signal = df[axis].values
    yf = np.abs(fft(signal))[:n // 2]  # 절반만 사용 (양의 주파수)
    xf = fftfreq(n, d=1/fs)[:n // 2]

    plt.plot(xf, yf, label=axis)

plt.title("FFT Spectrum (Amplitude vs Frequency)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def show_specto(file_path):
    # CSV 파일 불러오기
    df = pd.read_csv(file_path)
    
    for axis in ['gyro_x', 'gyro_y', 'gyro_z']:
        plt.figure(figsize=(6, 2))
        plt.specgram(df[axis].values, Fs=fs, NFFT=64, noverlap=32, cmap='viridis')
        plt.title(f"Spectrogram - {axis}, {file_path.split('/')[-1]}")
        plt.xlabel("Time (s)")
        plt.ylabel("Frequency (Hz)")
        plt.colorbar(label="Intensity (dB)")
        plt.tight_layout()
        
        plt.show()
```

```python
show_specto("./data/train/exp_1_DAY2/DAY2_exp1.csv")
```
